```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.info()
titanic.head()
```

```python
import matplotlib.cm as cm
x = titanic.sex.value_counts()
print(x)
labels = ['Males', 'Females']
```

```python
plt.pie(x, labels=labels, autopct='%1.1f%%', shadow=True)
```

```python
plt.figure(figsize=(18,5))
y = titanic.sex
gender_color=[]
for elem in y:
  if elem=="male":
    gender_color.append("green")
  else:
    gender_color.append("blue")

plt.scatter(titanic.fare, titanic.age, c=gender_color);
```
