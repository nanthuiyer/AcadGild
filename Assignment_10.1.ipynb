```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
df.head()
```

```python
#Delete unnamed columns
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1)
```

```python
#Show the distribution of male and female
by_gender = df.groupby('Gender')
by_gender.size().plot(kind='bar')
by_gender.size()
```

```python
#Show the top 5 most preferred names
most_preferred = df['Name'].value_counts()[df['Name'].value_counts() == df['Name'].value_counts()]
most_preferred[0:5]
```
